```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales = pd.read_csv('sales-canada.csv', index_col = 0)
sales
```

```python
sales = sales.apply(lambda x: x.str.strip('A'))
sales = sales.apply(lambda x: x.str.strip('B'))
sales = sales.apply(lambda x: x.str.strip('C'))
```

```python
sales
```

```python
sales.replace(['x', '..'], 0, inplace= True)
```

```python
sales = sales.apply(lambda x: x.str.replace(',',''))
```

```python
for i in range(len(sales.columns)):
    sales.iloc[:,i] = pd.to_numeric(sales.iloc[:,i])
```

```python
sales.dtypes
```

```python
plt.figure(figsize=(10,6))
for i in range(len(sales)):
    plt.plot(sales.columns[-4:], sales.iloc[i,-4:], label = sales.index[i])
    plt.xticks(rotation = 90)
    plt.legend(bbox_to_anchor=(1.1, 1.05))
plt.ylabel('Sales in Millions of Dollars')
plt.xlabel('Month-Year')
```

```python
plt.figure(figsize=(10,6))
for i in range(len(sales)-13):
    plt.plot(sales.columns[-12:], sales.iloc[i-13,-12:], label = sales.index[i-13])
    plt.xticks(rotation = 90)
    plt.legend(bbox_to_anchor=(1.1, 1.05))
plt.ylabel('Sales in Millions of Dollars')
plt.xlabel('Month-Year')
```

```python
plt.figure(figsize=(10,6))
for i in range(len(sales)-13):
    plt.plot(sales.columns[-12:], sales.iloc[i+13,-12:], label = sales.index[i+13])
    plt.xticks(rotation = 90)
    plt.legend(bbox_to_anchor=(1.1, 1.05))
plt.ylabel('Sales in Millions of Dollars')
plt.xlabel('Month-Year')
```

```python
cpi = pd.read_csv('CPI.csv', index_col=0)
cpi = cpi.iloc[:,36:]
cpi
```
