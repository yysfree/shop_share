```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import requests
```

```python
page_numbers = np.array([i for i in range(1,91)])
all_names = []
all_postals = []
all_addresses = []
```

```python
def clean(list, all_list):
  for element in list:
    element = str(element)
    element = element[element.find('">')+2:element.find('</')]
    all_list.append(element)
```

```python
links = []
for page_number in page_numbers:
  url = 'https://411.ca/business/search?q=accessories%20retail%20toronto%20ON&p=' + str(page_number)
  r = requests.get(url)
  html_doc = r.text
  soup = BeautifulSoup(html_doc)
  for div in soup.find_all('li'):
    a = div.find('span', {'itemprop':'streetAddress','class': 'ng-star-inserted'})
    name = div.find('h4', {'class': 'h6'})
    post = div.find('span', {'itemprop':'postalCode','class': 'ng-star-inserted'})
    if a is not None and post is not None:
      links.append(name)
      links.append(a)
      links.append(post)
```

```python
links[:5]
```

```python
length_links = len(links)
for i in range(length_links):
  element = str(links[i])
  if(i %3 == 0):
    element = element[element.find('">')+2:element.find('</')]
    all_names.append(element)
  elif(i%3 == 1):
    element = element[element.find('">')+2:element.find(',')]
    all_addresses.append(element)
  else:
    element = element[element.find('">')+2:element.find('</')]
    all_postals.append(element)

```

```python
final_dict = {'store_name': all_names, 'street_address': all_addresses, 'postal_code': all_postals}
```
